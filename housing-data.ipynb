```python
df = pd.read_csv('/kaggle/input/melbourne-housing-snapshot/melb_data.csv')
df.describe()
```

```python
y = df.Price
X = df.drop('Price', axis = 1)

X = X._get_numeric_data()

missing = [col for col in X.columns
               if X[col].isnull().any()]




X.head()
```

Below the imputer calculates the average value of each column and inserts it into the numerical values that are missing in the column, we then create another column to determines whether the imputer has been used or not 

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()

for col in missing:
    X[col + ' missing'] = X[col].isnull()
    

names = X.columns

ImputedX = pd.DataFrame(my_imputer.fit_transform(X))

ImputedX.columns = names

ImputedX.head()

```

```python
from xgboost import XGBRegressor

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)

model = XGBRegressor(n_estimators = 800)
model.fit(X_train,y_train)

predictions = model.predict(X_test)


from sklearn.metrics import mean_absolute_error

MAE = mean_absolute_error(predictions, y_test)

print(MAE)
```
